At each $t$, each user $k$ choose a channel $n$ to transmit. The action vector is $(a^k)_{k\in\{1,\dots,K\}}$ where $a^k \in [0, N]$ is the channel user $k$ chooses. $a^k=0$ means that the user $k$ chooses not to transmit.

The channel vector $\boldsymbol{c}_t \in \{0, 1\}^N$ evolves according to a bernouilli distribution.

If more than 1 user chooses the same channel there is a collision and the BS does not have any information about the channel and the packet. On the other hand, if a single user chooses a bad channel, the BS has the information that the channel is bad. The BS then in the ACK/NACK symbol, brodcasts the channel information it gathered to all users i.e. the vector $\boldsymbol{\hat{c}}_t = (\hat{c}^n_t)_{n=1,\dots N}$ where: 
- $\hat{c}^n_t = 1$ if the channel is good and used by a single user.
- $\hat{c}^n_t = d^n_t$ if the channel is good and a collision occur, where $d_n$ is inverse proportional to the number of users that have selected the channel. 
- $\hat{c}^n_t = 0$ if no one has selected the channel.
- $\hat{c}^n_t = -1$ if the channel is bad.

The reward can be:
- the sum of channel quality: $\sum_{n=1^N}\hat{c}^n_t$.
- the number of discarded packets (to determine if it is possible to access in practice).
- the throughput: $\sum_{n=1^N}1_{(\hat{c}^n_t=1)}$.

The state is the concatenation of the buffer matrix $\boldsymbol{B}_t$ and the channel vector $\boldsymbol{c}_t$: $\boldsymbol{s}_t = (\boldsymbol{B}_t, \boldsymbol{c}_t) \in {\mathbb{N}^{K \times \delta}} \times \{0, 1\}^N$.

The observation of a device $k$ is the concatenation of its buffer with the last channel observation from the BS: $o^k_t = (\boldsymbol{b}^k_t, \boldsymbol{\hat{c}}_{t-1})$. Its size is $\delta_k + N$.

```python
class RandomAccess:
    def __init__(self, env, verbose=False):
        self.env = env
        self.verbose = verbose
    
    def act(self, buffers):
        n_packets = buffers.reshape((self.env.n_agents, self.env.deadlines.max())).sum(1)
        actions = np.random.choice(np.arange(0, self.env.n_channels+1), size=self.env.n_agents)
        actions[n_packets == 0] = 0
        return actions
    
    def run(self, n_episodes):
        number_of_discarded = []
        number_of_received = []
        rewards_list = []
        jains_index = []
        channel_score = []
        for _ in range(n_episodes):
            rewards_episode = []
            done = False
            _, (buffer_state, channel_state) = self.env.reset()

            while not done:

                action = self.act(buffer_state)
                _, next_state, reward, done, _ = self.env.step(action)
                buffer_state = next_state[0]
                rewards_episode.append(reward)

            rewards_list.append(np.sum(rewards_episode))
            number_of_received.append(self.env.received_packets.sum())
            number_of_discarded.append(self.env.discarded_packets.sum())
            jains_index.append(self.env.compute_jains())
            channel_score.append(self.env.compute_channel_score())
            
            
        if self.verbose:
            print(f"Number of received packets: {np.sum(number_of_received)}")
            print(f"Channel score: {np.mean(channel_score)}")

        return 1 - np.sum(number_of_discarded) / np.sum(number_of_received), np.mean(jains_index), np.mean(channel_score), np.mean(rewards_list)
```

```python
def Capa(gamma):
    return np.log2(1+gamma)

def Disp(gamma):
    return gamma/2*(gamma+2)/(gamma+1)**2*np.log2(np.e)**2

def computeepsilon(gamma_db,L,n):
    gamma = 10**(gamma_db/10)
    epsilon = 1-norm.cdf(np.sqrt(n/Disp(gamma))*(Capa(gamma)-R))
    return epsilon

def pathloss(fc,d,hb,hd):
    d3D = np.sqrt(d**2+(hb-hd)**2)
    PLL = 32.4+17.3*np.log10(d3D)+20*np.log10(fc)
    PLN = 38.3* np.log10(d3D) + 17.3 + 24.9 * np.log10(fc)
    return np.maximum(PLL,PLN)
```

```python
class TestInit:
    def __init__(self):
        self.n_cells = 5
        self.users_per_cell = 10
        self.distance_between_bs = 300
        self.total_users = self.n_cells * self.users_per_cell
        self.radius = 150
        self.W = 38.16e6 # 40MHz minus guard bands
        self.pt = 10**(23/10)
        N_db = -174+10*np.log10(self.W)+NF
        self.N = 10**(N_db/10)
        pl = np.zeros((self.n_cells, self.total_users))
        shd = self.init_shadowing()
        self.G_BS = 5
        self.G_UE = pl + self.G_BS + shd
        self.max_simultaneous_devices = 3

    def init_shadowing(self):
        return np.random.normal(0, 8.02, (self.n_cells, self.total_users))
        
    def position_bs(self):
        side = int(np.ceil(self.n_cells**0.5))

        square_vertex = np.vstack(np.ones((side, side)).nonzero()).T
        square_vertex *= self.distance_between_bs
        
        points = square_vertex[:self.n_cells]
        self.bs_coordinates = points
    
    def compute_distance_matrix(self, A, B):
        num_A = A.shape[0]
        num_B = B.shape[0]
        
        A_square = np.sum(A ** 2, axis=1).reshape(num_A, 1)
        B_square = np.sum(B ** 2, axis=1).reshape(1, num_B)
        
        dot_product = np.dot(A, B.T)
        
        distances = np.sqrt(A_square + B_square - 2 * dot_product)
        
        return distances


    def generate_points_in_cell(self, coord, n, d):
        x, y = coord
        angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
        x_coords = x + d * np.cos(angles)
        y_coords = y + d * np.sin(angles)
        points = np.vstack([x_coords, y_coords]).T
        return np.array(points)
    
    def generate_users_positions(self):
        self.layout = {}
        self.distances_matrix = np.zeros((self.n_cells, self.total_users))
        for i, bs in enumerate(self.bs_coordinates):
            users = self.generate_points_in_cell(bs, self.users_per_cell, self.radius)
            self.layout[f"cell_{i}"] = users
            start = int(i*self.users_per_cell)
            end = int((i+1)*self.users_per_cell)
            self.distances_matrix[:, start:end] = self.compute_distance_matrix(self.bs_coordinates, users)

    
    def plot_layout(self):
        # Plot BS
        x = [point[0] for point in self.bs_coordinates]
        y = [point[1] for point in self.bs_coordinates]
        
        plt.scatter(x, y, color='blue', label='Base Station')
        
        for i in range(len(x)):
            circle = plt.Circle((x[i], y[i]), self.radius+1, fill=False, color='blue', linestyle='dashed')
            plt.gca().add_patch(circle)

        
        # Plot users
        users = np.concatenate(list(self.layout.values()))
        xu = [point[0] for point in users]
        yu = [point[1] for point in users]
        
        plt.scatter(xu, yu, color='green', label='Devices')
        
        plt.axhline(0, color='black', linewidth=0.5)
        plt.axvline(0, color='black', linewidth=0.5)
        
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Environment layout')
        plt.legend()
        plt.grid()
        
        plt.show()
        
    def decode_signal(self, attempts, pg, h_coeffs):
        h_coeffs = np.array(h_coeffs)#.reshape(self.n_cells, self.users_per_cell)
        attempts_matrix = attempts.reshape(self.n_cells, self.users_per_cell)
        n_attempts_per_cell = attempts_matrix.sum(1) # Transmission attempts per cell
        
        decoded_cells = np.zeros(self.n_cells)
        decoded_cells[np.where(n_attempts_per_cell > self.max_simultaneous_devices)] = -1 #Collision happened in the cell.

        for cell in range(self.n_cells):
            print(f"cell {cell}\n")
            decoded_idx = []
            pg_cell = pg[cell]
            active_idx = attempts.nonzero()[0]

            if n_attempts_per_cell[cell] <= self.max_simultaneous_devices:
                attempts_idx = attempts_matrix[cell].nonzero()[0] 
                # We add cell in order to have the absolute index (in total_users) and not relative to the cell.
                attempts_idx += cell * self.users_per_cell
                h2 = np.absolute(np.diag(h_coeffs))
                eta = pg_cell * h2.reshape(-1) # Shape: total_users
                eta_attempts = eta[attempts_idx]
                decoding_order = (-eta_attempts).argsort() # We put "-" because we want to sort by descending
                attempts_order = attempts_idx[decoding_order]
                # To have the real indices of the devices to decode: attempts_idx[decoding_order]
                
                print(f"active idx: {active_idx}")
                print(f"attempts: {attempts_idx}, eta: {eta}, decoding order: {attempts_idx[decoding_order]}")
                print(f"fast fading ||h||^2: {h2}")
                print(f"pg_cell: {pg_cell}")
                print(f"eta_attempts: {eta_attempts}")

                eps_list = []
                for device in attempts_order:
                    interference = np.delete(active_idx, decoded_idx + [device])
                    sinr = eta[device] / (self.N + (pg_cell[interference] * np.absolute(h_coeffs[device,
                                                                               interference])**2 / h2[device]).sum())
                    # sinr = eta_attempts[device] / (self.N + (eta_attempts[interference]).sum())

                    #eps = self.compute_epsilon(sinr, self.n_pilots)
                    #eps_list.append(eps)
                    #rv = np.random.binomial(1, 1-eps)
                    
                reward_sarl = len(decoded_idx)


        
        return decoded_idx

    def sample_multivariate(self, a):
        #size: (k, nt, 2)
        return np.random.multivariate_normal(np.array([0, 0]), (1-a**2)*0.5 * np.diag([1, 1]), size=(self.total_users, 4))


    def compute_channel_components(self):
        g = 10**(self.G_UE / 10) # g is a matrix where a component i,j is the gain of the user j for the BS i
        h = self.sample_multivariate(0)
        cmat = np.matrix(h[:,:, 0] + 1j*(h[:, :, 1]))
        h_coeffs = cmat.dot(cmat.getH())

        return self.pt * g, h_coeffs
```
